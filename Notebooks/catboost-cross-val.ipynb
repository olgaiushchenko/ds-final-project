```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 999)
```

```python
from catboost import Pool, CatBoostRegressor
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score
from sklearn.metrics import mean_absolute_error, median_absolute_error
from sklearn.model_selection import TimeSeriesSplit
```

```python
import feature_engineering_multi
```

Reading the dataset from csv file to pandas Dataframe object

```python
df = pd.read_csv('/Users/olia/galvanize/ds-final-project/data/Fire_Department_Calls_for_Service.csv')
```

Applying feature engineering script to create/reshape/upddate/drop features. Script is uploaded from outside file.

```python
df_full=feature_engineering_multi.run_features(df)
```

Splitting dataset into X dataframe and y labels.

```python
y = df_full['Respond to Available']
X = df_full.drop(columns=['Respond to Available'])
```

Creating 6 folds for cross validation.

```python
tscv = TimeSeriesSplit(n_splits=6)
```

Indeces of categorical features.

```python
cat_features=[
    0,
    1,
    3,
    5,
    7,
    9
]
```

```python
n = 1
mae_all =[]
mse_all = []
rmse_all = []
r2_score_all = []

for train, test in tscv.split(X):
    X_train, X_test = X[train[0]:train[-1]], X[test[0]:test[-1]]
    y_train, y_test = y[train[0]:train[-1]], y[test[0]:test[-1]]
    print('Loop #:', str(n))
    n +=1
    train_pool = Pool(X_train, y_train, cat_features=cat_features)
    test_pool = Pool(X_test, cat_features=cat_features)
    print('Pools created, fitting model now')
    model = CatBoostRegressor(iterations=100, depth=16,learning_rate=0.1, one_hot_max_size=3)
    model.fit(train_pool)

    print('Predicting now, almost there!')
    pred = model.predict(test_pool)
    pred[pred < 0] = 0 
    
    mae = mean_absolute_error(y_test,pred)
    mse = mean_squared_error(y_test,pred)
    rmse = model.score(test_pool, y_test)
    r2_score_val = r2_score(y_test,pred)
    
    mae_all.append(mae)
    mse_all.append(mse)
    rmse_all.append(rmse)
    r2_score_all.append(r2_score_val)
    
    print ('Mean absolute error: ', str(mae))
    print ('Mean squared error: ', str(mse))
    print ('Root squared eroor: ', str(rmse))
    print ('Score of determination: ', str(r2_score_val))
    print('-----new loop ------')
    
```
